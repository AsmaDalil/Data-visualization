## Task1: Calculating Area


```python
from shapely.geometry import Point

# Create a GeoSeries object
geometry = gpd.GeoSeries([Point(0, 0), Point(1, 1), Point(2, 2)])

# Access the area attribute
areas = geometry.area
print(areas)
```

#### Plotting the bar chart

```python
import matplotlib.pyplot as plt

# Sorting the data by area (optional but can help with visualization)
world_data_sorted = world_data.sort_values(by='area', ascending=False)

# Plotting the bar chart
plt.figure(figsize=(12, 8))
plt.bar(world_data_sorted['NAME'], world_data_sorted['area'], color='skyblue')
plt.xlabel('Country')
plt.ylabel('Area (square units)')
plt.title('Area of Each Country')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

## Removing a Continent


```python
# Removing Antarctica from GeoPandas GeoDataframe
world_data = world_data[world_data['NAME'] != 'Antarctica']
world_data.plot() 
```

#### Moving average chart 


```python
# Calculate the moving average
window_size = 5
moving_avg = world_data['area'].rolling(window=window_size).mean()

# Plotting the original data and moving average
plt.figure(figsize=(12, 8))
plt.plot(world_data.index, world_data['area'], label='Original Data', color='skyblue')
plt.plot(world_data.index, moving_avg, label=f'Moving Average ({window_size}-period)', color='orange')
plt.xlabel('Index')
plt.ylabel('Area (square units)')
plt.title('Area of Each Country with Moving Average')
plt.legend()
plt.show()
```

## 2nd task 

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt

# Calculate population density
world_data['population_density'] = world_data['POP2005'] / world_data['AREA']

# Define window size for moving average
window_size = 5

# Calculate the moving average of population density for each region
world_data['moving_avg_population_density'] = world_data.groupby('REGION')['population_density'].rolling(window=window_size, min_periods=1).mean().reset_index(level=0, drop=True)

# Plotting the moving average chart
plt.figure(figsize=(12, 8))
for region in world_data['REGION'].unique():
    region_data = world_data[world_data['REGION'] == region]
    plt.plot(region_data.index, region_data['moving_avg_population_density'], label=region)

plt.xlabel('Index')
plt.ylabel('Moving Average Population Density')
plt.title('Moving Average of Population Density Over Time for Different Regions')
plt.legend()
plt.show()
```

#### Plotting the area chart

I  calculate the cumulative population density for each region over time using the cumsum() function.
Then, I plot the area chart where each region is represented by a colored area, and the total area under the curve represents the cumulative population density.

```python
# Calculate cumulative population density for each region over time
cumulative_density = world_data.groupby('REGION')['population_density'].cumsum()

# Plotting the area chart
plt.figure(figsize=(12, 8))
for region in world_data['REGION'].unique():
    region_data = world_data[world_data['REGION'] == region]
    plt.fill_between(region_data.index, cumulative_density[region_data.index], label=region)

plt.xlabel('Index')
plt.ylabel('Cumulative Population Density')
plt.title('Cumulative Population Density Over Time for Different Regions')
plt.legend()
plt.show()
```
